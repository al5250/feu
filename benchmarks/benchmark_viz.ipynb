```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('./ssass.mplstyle')
```

```python
df = pd.read_csv('results.csv')
df.loc[8] = {'technique': 'FEU', 'nmi_mean': 0.977, 'nmi_std': 0.024, 'ari_mean': 0.973, 'ari_std': 0.030, 'acc_mean': 0.979, 'acc_std': 0.019}
df
```

```python
ax = df.plot(x='technique', y='nmi_mean', yerr='nmi_std', kind='bar', legend=False)
plt.xlabel('Technique')
plt.ylabel('NMI')
ax.patches[8].set_facecolor('blue')
plt.tight_layout()
plt.savefig('figures/nmi.png', dpi = 2000)
# save as pdf
plt.savefig('figures/nmi.pdf', format='pdf', dpi = 2000)
plt.show()
```

```python
ax = df.plot(x='technique', y='ari_mean', yerr='ari_std', kind='bar', legend=False)
plt.xlabel('Technique')
plt.ylabel('ARI')
ax.patches[8].set_facecolor('blue')
plt.tight_layout()
plt.savefig('figures/ari.png', dpi=2000)
# save as pdf
plt.savefig('figures/ari.pdf', format='pdf', dpi=2000)
plt.show()
```

```python
ax = df.plot(x='technique', y='acc_mean', yerr='acc_std', kind='bar', legend=False)
plt.xlabel('Technique')
plt.ylabel('Accuracy')
ax.patches[8].set_facecolor('blue')
plt.tight_layout()
plt.savefig('figures/acc.png', dpi = 2000)
# save as pdf
plt.savefig('figures/acc.pdf', format='pdf', dpi = 2000)
plt.show()
```
